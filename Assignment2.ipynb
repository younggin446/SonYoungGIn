#### 2번

경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.

인자로 전달하는 문자열은 경로만 들어간다고 가정한다.

각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.

반환하는 리스트의 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.

다음과 같은 실행 및 출력 결과가 도출되어야 한다.
div_path('/usr/local/bin/python')
['/usr/local/bin', 'python']
div_path('/home/chulsoo/test.txt')
['/home/chulsoo', 'test.txt']

[참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.

```python
def div_path(address) :
    list1 = [] # 문자 하나씩 모두 저장할 배열 list1을 생성 및 초기화
    
    for s in address.strip() : # 주소 앞, 뒤에 공백 있을 시 strip()으로 없애고 for 반복문 실행
        list1.append(s) # list1 배열에 문자열 하나씩 모두 삽입
    
    s = "".join(list1) # 배열 안 하나의 문자열씩 저장되어있는 문자들을 join 메서드를 이용하여 공백없이 이어주기 
    l = s.split("/") # 이렇게 이어준 s 문자열들을 "/"을 기준으로 분리하기
    
    dir_list = [] # 디렉토리를 저장할 배열
    file_list = [] # 파일명을 저장할 배열
    df_list = [] # 디렉토리와 파일명을 따로 저장할 배열
    
    for i in l[:len(l) - 1] : # 파일명 전까지 즉 디렉토리명을 dir_list 배열에 저장
        dir_list.append(i) 
    l2 = "/".join(dir_list) # join 메서드를 이용하여 각 디렉토리를 "/"로 연결하기
    
    for i in l[len(l) - 1:] : # 파일명을 file_list 배열에 저장
        file_list.append(i)  # 디렉토리를 제외한 나머지 파일명을 file_list 배열에 저장
    l3 = "".join(file_list)
    
    # 위의 디렉토리명과 파일명을 df_list배열 각 인덱스에 저장
    df_list.append(l2) 
    df_list.append(l3)
    
    print(df_list)
    
# 실행 
div_path('/usr/local/bin/python')
div_path('/home/chulsoo/test.txt')
```

#### 설명

이 문제는 파일명과 디렉토리명을 각각 따로 분리하는 함수를 작성하는 문제입니다.
먼저 이 함수의 이름은 문제에서 제시한 대로 div_path로 하고 매개변수로 주소를 전달받습니다. 먼저 이 전달받은 주소의 각각의 문자열 모두를 list1이라는 배열에 저장하기 위해 함수 바로 아래부분에서 선언한 list1배열에 for 반복문과 append메소드를 이용하여 삽입합니다. 단 주소의 앞,뒤에 공백이 있을 수 있으므로 strip메소드를 이용합니다.

이제 배열 안에 하나의 문자씩 저장되어 있기 때문에 join 메서드를 이용하여 공백으로 이어준 뒤 s변수에 저장하고, "/"을 기준으로 분리합니다.
"/"으로 분리하는 이유는 디렉토리명과 파일명을 분리하기 위함입니다. 디렉토리명은 "/"을 함께 출력해야 하는데 이는 뒤의 for문에서 다시 출력하면 되기 때문에 일단 "/"을 기준으로 분리합니다.

이제 디렉토리를 저장할 배열(dir_list), 파일명을 저장할 배열(file_list) 그리고 디렉토리와 파일명을 각각 저장할 배열(df_list)을 생성하고 빈 리스트로 초기화합니다. 그리고 for문을 이용하여 각각 저장을 합니다. 일단 파일명 전까지는 모두 디렉토리명이므로 길이를 이용하여 위에서 "/"으로 분리한 l 배열의 총 길이 전까지 즉, 디렉토리명을 dir_list배열에 삽입합니다. for문이 모두 끝나고 나면 join 메서드를 이용하여 각각 디렉토리 사이에 "/"을 삽입하여 l2변수에 저장합니다.
디렉토리를 제외한 나머지는 파일명으로 총 길이 - 1부터 끝까지 즉, 파일명을 file_list배열에 삽입합니다. 그리고 이 삽입한 파일명 리스트를 l3 변수에 저장합니다.
위에서 저장한 l2와 l3를 디렉토리와 파일명을 각각 저장할 배열(df_list)에 삽입하여 출력해주면 이 문제는 해결입니다!

#### 3번 

두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.

집합 자료형 set은 사용하지 않는다.

인자로 전달하는 리스트 2 개에는 정수 값만 들어간다고 가정하자.

함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.

반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).

반환하는 리스트는 정렬이 되어 있어야 한다.

다음과 같은 실행 및 출력 결과가 도출되어야 한다.

※ list_union([1, 2, 3], [1, 2, 4])

[1, 2, 3, 4]

※ list_union([-10, -5, 0, -1], [100, 9, 0, 9])

[-10, -5, -1, 0, 9, 100]

※ list_union([0, 1, 2], [0, 1, 2])

[0, 1, 2]

[참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.

[참고] 임의의 정수값 (x)이 리스트 (l) 내에 존재하는지 판단하는 방법은 x in l 이다.

```python
def list_union(lista, listb) :
    for i in listb :
        if i not in lista :
            lista.append(i)
            
    new_list = []
    for i in lista :
        if i not in new_list :
            new_list.append(i)
            
    return sorted(new_list)
    
print(list_union([1, 2, 3], [1, 2, 3, 4]))
print(list_union([-10, -5, 0, -1], [100, 9, 0, 9]))
print(list_union([0, 1, 2], [0, 1, 2]))
```

#### 설명

※ 합집합(A U B)
문제에서 제시한 list_union 함수는 lista와 listb 두 개의 리스트를 매개변수로 받습니다. 그리고 for문을 이용하여 listb에 있는 요소들을 하나씩 if문으로 검사를 합니다. 이 if문은 listb에 있는 요소들이 lista에 있는 값인지 아닌지를 not과 in을 통해 검사하고 만약 lista에 존재하지 않는 요소이면 append 메소드를 이용하여 lista 배열에 삽입합니다.
그리고 문제에서 중복된 요소들의 출력은 1번만 출력하고 set 자료형은 사용하지 않을 것을 요구했으므로 중첩 반복문을 통해 반복된 요소들의 값을 제거합니다. 방법은 new_list라는 이름의 새로운 리스트를 만들고 lista 배열에 있는 값들을 not과 in을 통해 검사하여 new_list 배열에 존재하지 않으면 append를 삽입합니다.
마지막으로 문제에서 요구한대로 정렬된 상태로 반환해야하기 때문에 sorted메소드를 이용하여 new_list 배열을 반환하면 문제 해결입니다!

#### 5번

두 개의 양의 정수를 인자로 받아서 해당 범위안에 있는 소수(prime number)의 리스트를 출력하는 함수 print_primenumber(a, b)를 작성하시오.

이미 구현된 모듈이나 라이브러리는 사용하지 않는다.

함수의 반환(return)은 없으며 함수 내의 끝에서 .format()을 사용하여 다음과 같이 출력한다.

print_primenumber(2, 10)

2 이상 10 이하 정수 중 소수 리스트: [2, 3, 5, 7]

```python
def print_primenumber(a, b) :
    
    prime_list = []
    
    for i in range(a, b + 1) :
        flag = True
        for j in range(2, i) :
            if(i % j == 0) :
                flag = False
                break
                
        if(flag == True):
            prime_list.append(i)
            
    print("{}이상 {}이하의 정수 중 소수 리스트: {}".format(a, b, prime_list))
    
print_primenumber(2, 10)
```

#### 설명

문제에서 두 개의 양의 정수를 매개변수로 받는다고 했으므로, a와 b를 매개변수로 받은 뒤
소수를 저장할 리스트 prime_list를 선언하고 빈 리스트로 초기화 합니다.
그리고 중첩 반복문을 이용하여 소수를 판별합니다. 중첩 반복문에서 첫 번째 for문의 범위는 (a, b + 1)로 설정합니다. 그 이유는 range(A, B)가 있다고 가정 시 정수는 A <= n < B로

웹 URL로 지정된 웹페이지를 문자열로 가져와 
1) 모든 HTML 태그를 제외한 순수 텍스트 문자열만을 걸러내고, 

2) 그 순수 텍스트 문자열 안에 존재하는 단어가 총 몇개인지 

출력하는 프로그램을 작성하시오

```python
import requests
req = requests.get('https://www.naver.com/')
source = req.text

source = source.strip() # source 코드의 앞 뒤 공백 제거
exceptHTML_list = [] # HTML 태그 제외한 문자열들을 저장할 배열
word_flag = True # "<"와 ">" 사이의 문자열들을 제외한 나머지 문자열들을 exceptHTML_list 배열에 삽입하기 위해 필요한 flag 변수

for ch in source :
    if (ch == "<") :
        word_flag = False
        continue
        
    elif (ch == ">" and word_flag == False) :
        word_flag = True
        continue
        
    if(word_flag) :
        exceptHTML_list.append(ch)

exceptHTML_list = ("".join(exceptHTML_list)).split()
print(len(exceptHTML_list)) 

real_word = [] # 순수 문자열들만 저장할 배열(영어, 한글, 한자)
for word in exceptHTML_list :
    if not word.isalnum() : continue #  모든 문자가 글자와 숫자가 아니라면 제외!
    
    elif not word.isalpha() : continue # 모든 문자가 알파벳 글자가 아니라면 제외!
    
    else : real_word.append(word) # 순수 문자열(영어, 한글, 한자)만 real_word 배열에 삽입
        
print(len(real_word))
```

#### 설명

먼저 requests 모듈을 사용하기 위해 requests를 import합니다. 그리고 제가 검사하고자 하는 네이버를 get메서드를 통해 가져온 뒤 txt 형식으로 source 변수에 저장합니다. 먼저 본격적으로 문자열 개수를 세기 전 공백은 문자열이 아니기 때문에 strip 메서드를 통해 source 앞 뒤에 있는 공백을 제거합니다. 그리고 HTML을 제외한 문자열들을 저장할 배열(exceptHTML_list)과 "<", ">"사이에 있는 문자열인지 검사하기 위해 flag 변수(word_flag)도 함께 선언합니다.
이제 반복문을 통해 "<"와 ">" 사이에 있는 문자열을 제외한 나머지 문자열을 위에서 선언한 exceptHTML_list에 삽입합니다. 이 때 총 3가지의 조건문을 사용합니다. 첫 번째 조건문에서는 source 코드 안의 문자열(ch)이 "<"이면 flag변수를 False로 초기화 하고, continue를 통해 아래의 코드가 아닌 다시 반복문의 처음으로 돌아갑니다. 그리고 두 번째 조건문에서 만약 source 코드 안의 문자열(ch)이 ">"이고 flag 변수가 False이면 다시 반복문의 처음으로 돌아갑니다. 이렇게 두 가지의 조건식 때문에 <>안에 있는 문자열들은 flag 변수가 False입니다. 따라서 마지막 조건문을 통해 flag변수가 True인 즉, <>안을 제외한 나머지 문자열들을 HTML을 제외한 나머지 문자열 배열에 삽입합니다.
그리고 join메서드를 통해 문자열들을 이어주고, split 메서드를 통해 공백을 기준으로 다시 배열에 삽입합니다. 이렇게 되면 <>을 제외한 나머지 문자열들이 exceptHTML_list 배열의 각 인덱스에 위치합니다. 마지막으로 len함수를 통해 배열 속 요소들의 개수를 세어주면 HTML 태그 제외한 문자열(단어)의 개수를 알 수 있습니다!
이렇게 문제를 풀다가 순수 문자열들만 저장할 수 있는 배열이 있을까에 대해 고민했습니다. 예를 들어 영어나 한글, 한자, 숫자 + 한/영 등과 같은 특수문자를 제외한 순수 문자열들의 개수를 셀 방법을 고민해보았습니다.
그래서 고민해본 결과 저번 주차에서 학습했었던 isalnum을 이용하여 모든 문자가 글자와 숫자가 아니면 제외하고, isalpha를 통해 알파벳 글자가 아니여도 continue를 통해 제외했습니다.
위에서 순수 문자열이라고 생각되지 않는 단어들을 제외한 나머지 순수 단어들을 real_word배열에 append메서드를 이용하여 삽입하고 len함수를 통해 길이를 출력하면 영어나 한글, 한자, 숫자 + 한/영 등과 같은 특수문자를 제외한 순수 문자열들의 개수 또한 알 수 있었습니다!
