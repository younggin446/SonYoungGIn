#### 2번문제

문자열 S = 'Hello World and Python'에 대해 다음 요구사항에 대한 Python 코드를 제시하시오.

1) 단어의 순서가 역순으로 되어진 문자열 ('Python and World Hello')을 만들고

2) 1)의 결과 문자열에 대해 메소드 split과 join을 이용하여 공백을 모두 없엔 문자열 'PythonandWorldHello'을 만드시오.

```python
S = "Hello World and Python"
reversedStr = ""

# 1)
word_list = S.split()
reversed_word = word_list[::-1]

for i in range(len(reversed_word)) :
    if(i != len(reversed_word) - 1) : reversedStr += (reversed_word[i] + " ")
    else : reversedStr += reversed_word[i]
        
print(reversedStr)

# 2)
no_blank_str = ''.join(reversedStr.split())
print(no_blank_str)
```

※ 설명

#### 1)
먼저 단어의 순서가 역순으로 되어진 문자열을 저장할 reversedStr 변수를 생성합니다. 그리고 빈 문자열로 초기화를 시킵니다.
먼저 역순으로 이루어진 문자열이 아닌 단어의 순서가 반대로 이루어진 문자열을 출력 해야하므로 각 단어별로 리스트에 저장을 합니다. 이때 split() 함수를 이용하여 공백을 기준으로 각 단어를 word_list 배열에 저장을합니다. 그리고 슬라이싱을 이용하여 각 단어들의 요소들의 순서를 반대로 바꾼 뒤 reversed_word에 저장을 합니다. 이제 단어들의 순서가 역순으로 이루어진 리스트가 생성이 되었습니다.

이제 for 반복문을 이용하여 출력을 합니다. 만약 i가 역순으로 이루어진 리스트의 길이의 -1과 같지 않다면 즉, 마지막 단어가 아니라면 단어와 단어 사이에 띄어쓰기를 함께 출력하기 위해 if문을 이용하여 단어와 공백을 같이 reversedStr 객체에 저장을 합니다. 그리고 마지막 단어 뒤에는 공백을 출력할 필요가 없으므로 else문을 통해서 단어만 저장을 합니다.
이렇게 단어의 순서가 반대로 이루어진 문자열 객체 reversedStr을 출력하면 문제 해결입니다.

#### 4번문제

다음 문자열을 ':'을 기준으로 분리하여 리스트로 만들고 각 문자열의 좌우 공백을 제거하시오 (즉, 문자열 S에서 l을 만들어라)

s = '  first star   :   second star   :    third star  '

l = ['first star', 'second star', 'third star']

1) for ~ in 문을 이용한 코드를 작성하시오.

2) 리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

```python
s = '  first star   :   second star   :    third star  '

# 1)
l = s.strip() # 앞 뒤 공백 제거
l = s.split(":") # ':'을 기준으로 분리 → 하지만 아직도 공백이 남아 있음
# 남아있는 공백을 for문을 통해 제거
for i in range(len(l)) :
    l[i] = l[i].strip()
    
print(l)
```

```python
# 2)
s = '  first star   :   second star   :    third star  '
l2 = s.split(":")
l2 = [l2[i].strip() for i in range(len(l2))]

print(l2)
```

하지만 아직도 공백이 각 문자열의 앞 또는 뒤에 남아 있으므로 for문을 이용하여 공백을 제거합니다. 각각의 인덱스의 요소들을 공백을 제거한 요소들로 다시 초기화하여 출력해보면 올바른 답이 나옴을 확인할 수 있습니다.

#### 5번 문제
다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.

1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

addall([1]) >>> 1

addall([1, 2, 3, 4, 5, 6, 7, 8, 9]) >>> 45

2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

addallodd([1]) >>> 1

addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]) >>> 25

[참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)

[참고 2]: sum() 등의 내장함수 사용 가능함

```python
# 1)
def addall(list) :
    sum = 0
    
    for i in list :
        sum += i
        
    return sum

print(addall([1]))
print(addall([1,2,3,4,5,6,7,8,9]))

def addAll(list) : return sum([i for i in list], 0)

print(addAll([1]))
print(addAll([1,2,3,4,5,6,7,8,9]))
```

※ 설명

설명하기에 앞서 [1]과 [1,2,3,4,5,6,7,8,9]가 각 함수(for문 이용 함수, 리스트 내포 이용 함수)에 잘 동작되는지 확인하기 위해 for ~ in 리터럴을 이용한 함수의 이름을 addall로 하였고, 리스트 내포 방식을 이용한 함수의 이름은 addAll로 지정하였습니다.

먼저 for문을 이용한 addall 함수부터 설명하겠습니다. 리스트 내의 모든 정수를 저장할 변수 sum을 0으로 초기화시킵니다. 그리고 for문을 이용하여 매개변수로 받아온 리스트의 각 요소들을 sum에 중첩하여 더합니다. 그리고 리스트 내의 요소들의 합이 저장된 sum을 리턴하면 문제가 해결됩니다.

두 번째 리스트 내포함수를 이용한 addAll함수입니다. 리턴 값으로 리스트 내포 함수를 1줄로 작성합니다. 이때 sum 내장 함수를 이용합니다. 주의할 점은 sum(iterable, start = 0)에서 start 부분입니다. sum 메서드의 초기 값을 지정해주지 않았으므로 start 부분 즉, 초기값을 0으로 설정하고 리스트 내포 함수를 작성합니다.

```python
# 2)

def addallodd(list) :
    sum = 0
    
    for i in list :
        if (i % 2 != 0) : sum += i
        
    return sum

print(addallodd([1]))
print(addallodd([1,2,3,4,5,6,7,8,9]))

def addAllodd(list) : return sum([i for i in list if i % 2], 0)

print(addAllodd([1]))
print(addAllodd([1,2,3,4,5,6,7,8,9]))
```

※ 설명

설명하기에 앞서 [1]과 [1,2,3,4,5,6,7,8,9]가 각 함수(for문 이용 함수, 리스트 내포 이용 함수)에 잘 동작되는지 확인하기 위해 for ~ in 리터럴을 이용한 함수의 이름을 addallodd로 하였고, 리스트 내포 방식을 이용한 함수의 이름은 addAllodd로 지정하였습니다.

먼저 for문을 이용한 addallodd 함수부터 설명하겠습니다. 리스트 내의 모든 정수를 저장할 변수 sum을 0으로 초기화시킵니다. 그리고 for문을 이용하여 매개변수로 받아온 리스트의 각 요소들을 sum에 중첩하여 더합니다. 단 홀수만 더해줘야 하기 때문에 if문을 이용하여 2로 나누었을 때 0이 아닌 즉, 짝수가 아닌 홀수들을 sum에 중첩하여 더합니다. 그리고 이 홀수들의 합이 저장된 sum을 리턴하면 문제가 해결됩니다.

length(X)를 다음과 같이 정의한다 

▶ 10진수로 표현된 X의 자릿수

Strange라는 숫자를 이렇게 정의한다. 

▶ 음이 아닌 정수 중 자릿수가 1인 모든 숫자는 Strange이다.

▶ length(X) > 1 인 숫자 X 는 다음 2가지를 만족하면 Strange이다.
X는 length(X)로 정확히 나누어 떨어진다.
X / length(X) 가 strange이다.

이 문제에서 숫자 하나가 주어질 때, 이 숫자가 Strange인지 판단하는 함수 isStrange(num)을 작성하시오.
isStrange(num) 함수가 재귀적인 함수가 되도록 작성해 보세요.

```python
def isStrange(num):
    num_digit = len(str(int(num))) # num 숫자의 자릿수
    
    if(num_digit == 1) : return True

    elif(num_digit > 1) :
        if (int(num % num_digit) == 0) :
            return isStrange(int(num / num_digit)) 
        else :
            return False
    else :
        return False
```

```python
print(isStrange(0))
print(isStrange(56))
print(isStrange(100))
print(isStrange(7882602506492313600))
print(isStrange(6012302549522098651))
```

또한, 이들은 그룹의 번호순으로 여행을 하길 원한다. 예를 들어, 3번 그룹은 1번 그룹과 2번 그룹이 여행을 한 뒤에 버스를 탈 수 있거나, 1번 그룹과 2번 그룹이 이미 버스에 탑승한 상태에 있을 때 동시에 버스를 탈 수 있다.

```python
def solve(a):
    
    all_passenger = 0 # 모든 승객 수
    
    for i in range(len(a)) : 
        all_passenger += a[i]
        
    take_bus = []
    
    for i in range(1, all_passenger + 1) :
        if (all_passenger % i == 0) :
            take_bus.append(i) # 1, 2, 3, 4, 6, 12
#------------------------------------------------------
    can_take_bus = [] 
    accmulate_passenger = 0 # 누적 승객 수

    for i in range(len(take_bus)) : 
        passenger_flag = True
        accumulate_passenger = 0
        
        for j in range(len(a)) :
            
            if (a[j] - take_bus[i] == 0) :
                continue
            
            elif(a[j] - take_bus[i] > 0) :
                passenger_flag = False
                break
                
            else :
                accumulate_passenger += a[j]
                continue
                
        if(passenger_flag) :
            can_take_bus.append(take_bus[i])

    return can_take_bus
```

※ 설명

문제에서 요구한 JavaScript와 CSS의 내용을 모두 삭제해야 합니다.
JavaScript : <script> ... </script>
CSS : <style> ... </style>

먼저 Javascript에 해당하는 <script> ... </script>를 지우는 코드부터 설명하겠습니다. if문을 이용하여 Javascript의 내용을 삭제할 때 슬라이싱을 이용하기 때문에 리스트의 범위에 벗어나는 오류를 발생시키지 않기 위해 최상위 반복문의 i를 source코드의 길이 - 1 이하일 때 코드가 실행되도록 했습니다.
이렇게 조건을 작성한 뒤, if문을 이용하여 <script>문자열인지 확인합니다. 확인 방법으로는 슬라이싱을 이용합니다. 먼저 <script>에서 "<"에 해당하는 지 확인하기 위해서 source 코드의 i번째 인덱스가 "<"인지, 그리고 다음 문자가 script인지 확인하기 위해서 source 코드의 i+1번째 인덱스 부터 i + 7번째 인덱스까지가 "script"인지 검사합니다.
i + 1부터 i + 7까지를 범위로 지정한 이유는 script는 6글자이기 때문에 위와 같이 범위를 지정하였습니다. 그리고 추가적으로 주의할 점은 script가 대문자일 수도 있기 때문에 lower()를 메서드를 이용합니다.
이제 <script>인지 확인이 끝났고, 문자열 "<script>" 검사가 끝난 마지막 인덱스 부터 문자열 "</script>"이 나오는지 검사해야하므로 마지막 문자열에 해당하는 ">"의 인덱스 즉 i + 7을 j에 초기화합니다. 
그리고 while문을 이용하여 </script>가 나올 때까지 검사를 해야 합니다. 검사를 하는 조건문은 위와 같은 동일하고 "/"부분을 검사하는 것만 and 연산자를 이용하여 추가합니다. 역시 슬라이싱 부분도 범위에 주의하여 코드를 작성합니다. 이렇게 if문을 작성한 뒤 j는 1씩 증가하는 코드를 작성합니다. 위의 문자열 "</script>"코드를 찾기 위해 인덱스를 계속 1씩 증가시키기 위한 코드입니다. 마지막으로 j 또한 source코드의 범위를 벗어나면 안되기 때문에 if문을 통해 조건에 벗어나면 break합니다.
CSS도 위와 같은 방식으로 슬라이싱 범위에 주의하여 동일하게 코드를 작성합니다.
    
HTML태그를 제거하는 코드는 저번주에 적은 코드와 동일하게 작성합니다. 그리고 이제 구두문자를 제거하기 위한 코드를 설명하겠습니다. 먼저 문자열 양 끝에 존재하는 구두문자를 없애기 위해서 for반복문을 이용합니다. 범위는 HTML 태그를 제거한 단어들을 삽입해놓은 리스트 exceptHTML_list의 길이만큼으로 지정합니다. 그리고 단어 하나하나 strip()메서드를 이용하여 문자열 양 끝 구두문자를 삭제합니다. 이제 문자열 내에 있는 구두문자를 삭제합니다. 이때도 역시 for문을 이용합니다. 이중 for문을 통해서 구두문자 즉, punctuation을 검사합니다. 이때 in연산자를 이용하여 검사를 하고, 만약 구두문자가 발견되면 구두문자에 해당하는 인덱스를 num에 저장합니다. num에 저장하는 이유는 슬라이싱을 통해서 구두문자를 제거한 문자열을 리스트에 삽입하기 위함입니다.이렇게 슬라이싱을 통해서 문자열 내에 있는 구두문자도 없애줍니다.
    
이제 리스트내에는 구두문자가 없는 단어들만 존재합니다. 하지만 문자열 끝에 있는 구두 문자를 제거할 때 공백 요소가 존재하게 됩니다. 이 리스트 내에 있는 공백을 삭제하기 위해서 리스트 내포함수를 이용합니다. 리스트 내포함수를 통해서 exceptHTML_list 리스트 내에있는 단어들이 공백이 아닌 단어에 해당하면 리스트에 삽입합니다.
    
이제 문제에서 원하는 마지막 요구사항은 사전을 생성하고 각 단어들의 개수를 출력하는 것입니다. 
먼저 dic이라는 dict 타입의 공백 dict를 생성합니다. 그리고 for문을 이용하여 key에 해당하는 단어들의 값(개수)을 0으로 초기화합니다. 이제 반복문을 통해서 exceptHTML_list내에 단어들이 위에서 지정한 key들과 동일한지 검사합니다. if문을 이용하여 검사를 하고 만약 동일한 단어(key)가 존재 시 1을 증가시킵니다. 마지막으로 dic을 출력 해주면 문제 해결입니다!
