#### 5번 문제

이 문제를 해결하기 위해서는 파이썬에서의 for문의 작성방법에 대해 알아야 합니다.
먼저 기존에 학습했던 C와 C++과는 유사하지만 in range()부분이 조금 다른 것을 확인할 수 있습니다. for i in range()에서 range()는 범위로, range(a, b)는 a 이상 b 미만만큼 반복을 의미합니다. 수식으로 작성하면 a <= 반복 횟수 < b입니다. 슬라이싱과 마찬가지로 b - 1까지만 반복되는 것에 주의합니다. 그리고 in 연산자를 사용하여 for문을 작성하는 것도 주의합니다!
이러한 for 반복문을 이용하여 홀수를 출력해야 합니다. 홀수는 2로 나누었을 때 나누어 떨어지지 않는 수이기 때문에, 조건문 속의 조건에 위와 같은 조건을 작성하고 출력을 합니다.
보기좋게 출력하기 위해 문자열 끝에는 공백으로 출력하기 위해 위와 같이 end = " "를 추가적으로 작성하였습니다!

```python
i = 1
sum = 0

while(i <= 100) :
    if(i % 2 == 0) :
        sum += i
        
    i += 1
    
print("1부터 100사이 짝수들의 합 : ", sum)
```

#### 7번 문제

이 문제는 위 3번 문제와 굉장히 유사한 문제로 역시, 슬라이싱의 개념을 이해해야 해결할 수 있는 문제입니다.
먼저 사용자에게 정수를 입력을 받습니다. 이때 input()을 통해 입력을 받고 사용자에게 무엇을 작성하는지 알려주는 프롬프트 문구도 함께 작성합니다. 그리고 여기서 주의해야 할 점은 input은 무조건 문자열로 입력을 받기 때문에 사용자가 정수를 입력을 했다하더라도 문자열로 저장이 됩니다. 따라서 슬라이싱을 사용할 수 있는 것입니다.
그리고 이제 입력한 정수형 문자열을 역순으로 출력하기 위해 슬라이싱을 사용하고, 문자열의 처음과 끝은 공백으로 작성 시 자동으로 입력이 되기 때문에 작성할 필요가 없으며, step부분을 -1로 해주면 역순으로 출력하는데 문제가 없음을 확인할 수 있습니다!

```python
num = int(input("정수를 입력하세요 : "))

for i in range(2, num + 1) :
    flag = True
    for k in range(2, i) :
        if(i % k == 0) : 
            flag = False
    
    if(flag) : print(i, " ")
```

#### 8번 문제

먼저 이 문제를 풀기 전 소수(Prime number)의 개념을 알아야 합니다. 소수는 약수가 자기자신과 1밖에 존재하지 않는 수로 오로지 약수의 개수가 2인 수를 의미합니다.
이렇게 소수에 대해 파악을 한 뒤 본격적으로 문제를 해결합니다. 먼저 사용자로부터 입력을 받아야 하기 때문에 input()을 통해 입력을 받습니다. 역시 프롬프트 문구도 함께 작성을 합니다. 여기서 주의할 점은 input()은 항상 문자열로 저장을 합니다. 근데 아래의 반복문에서 % 계산(나누기)을 해야하는데 문자열일 경우 계산을 할 수가 없습니다. 따라서 문자열을 정수 형태로 저장을 하기 위해 int()를 통해 한번 더 묶어줍니다.
이제 반복문을 통해 소수를 가려야 합니다. 여기서 반복문을 2번 즉, 중첩 반복문을 사용해야 합니다. 그 이유는 소수인지 아닌지 판별해야 하는 수는 1과 자기자신의 수를 제외한 나머지 수로 모두 나누어 보면서 나누어지면 소수에서 제외를 시켜야 하기 때문입니다. 즉 첫번째 반복문에서 i는 검사할 정수, 두번째 반복문의 k는 나눌 수를 의미합니다.
주의할 점은 위에서 사용자가 입력한 정수 사이에 존재하는 소수를 요구하고 있으므로 첫번째 for문의 범위에 사용자가 입력한 숫자에 +1을 해줍니다. 이유는 range()의 범위 때문입니다.
이어서 소수임을 구분하기 위해 불 타입의 flag변수를 이용했습니다. 이 플래그 변수의 위치도 역시 주의해야하는데, flag변수는 첫번째와 두번째 반복문 사이에 위치해야 합니다. 어떤 수가 소수인지 아닌지 True 또는 False로 초기화가 이루어진 후 다시 flag변수를 True로 초기화 시켜 검사해야하기 때문입니다. 이 flag변수를 처음에 True로 초기화를 시켜놓고 두 번째 반복문에서 만약 2부터 자기자신 이전의 수들로 나누어보면서, 나누어 지면 소수가 아니기 때문에 flag를 False로 초기화시킵니다.
1부터 사용자가 입력한 정수 사이의 소수인 수는 flag가 True로 초기화가 되어있기 때문에 중첩 반복문을 빠져나간 뒤 조건문을 통해 flag가 True인 경우 즉, 소수인 경우만 출력을 하면 문제를 해결할 수 있습니다!

```python
s = input("String S를 입력하세요 : ")

StuartScore = 0 # Stuart의 점수를 저장할 변수
KevinScore = 0 # Kevin의 점수를 저장할 변수

while(True) : # 무한반복
    StuartAnswer = input("자음으로 시작하는 단어 입력 : ")
    KevinAnswer = input("모음으로 시작하는 단어 입력 : ")
    
    n = 1
    m = 0
    count = 0
    
    for i in range(0, len(s)) : # 문자열의 길이만큼 반복
            
        for k in range(0, len(s) - count) : 
            
            if (StuartAnswer == s[m:m+n]) : StuartScore += 1
            if (KevinAnswer == s[m:m+n]) : KevinScore += 1
            
            m += 1
            
        n += 1
        count += 1
        m = 0
        k = 0
        
        # 문자열의 길이만큼 반복을 하고, 중첩 반복문 안에서는 문자열의 길이 ~ 1까지를 반복 → count 사용
        # BANANA를 예시로 들면 이해가 됨.
        # 중첩 반복문을 1회 끝나고 나왔을 때  다시 입력한 문자열의 0번째부터 검사를 해야하므로 m, k는 0!
        # n은 검사할 문자열의 길이로 1번째 반복문에서는 1개의 문자, 두번째 반복문 일 때는 2개의 문자,... n개의 문자는 n개의 문자로 +1씩 증가
    
    if(StuartAnswer == s or KevinAnswer == s) : break
    
if(StuartScore > KevinScore) :
    print("승자 : Stuart 점수 : ", StuartScore)
else :
    print("승자 : Kevin 점수 : ", KevinScore)
```

먼저 처음에도 인덱스 0부터 문자열을 검사하므로 m은 0, 마찬가지로 처음에는 1글자부터 비교를 하므로 n = 1로 while문에서 각각 초기화합니다. 그리고 두 번째 반복문이 끝나면 m을 0으로 초기화를 하여 항상 0번째 인덱스부터 검사를 하게 했으며 n은 1글자, 2글자, ... n글자를 검사를 해야 하기 때문에 두번째 반복문이 끝나면 +1을 해주어 글자 수에 맞는 검사를 하게 했습니다.
이러한 규칙을 작성한 코드를 이용하여 if문으로 Stuart가 입력한 문자열 또는 Kevin이 입력한 문자열과 같은지 검사를 합니다! 그리고 입력한 문자열이 있다면 중첩 연산자(+=)를 이용하여 +1점을 합니다.

##### 나머지 코드 설명
이 게임은 더 이상 입력할 단어가 없으면, 즉 둘 중 하나라도 정답을 작성하면 게임이 끝나야 하므로 while문의 마지막 부분에 if문을 이용해 진행자가 입력한 문자열 s(== 정답)와 Kevin 또는 Stuart가 입력한 문자열이 같은지 비교를 하고, 같다면 break를 통해 while문을 빠져나갑니다.
이렇게 빠져나오고 나서 승자가 누구이고 승자의 점수도 출력을 해야하므로, if - else문을 이용해 승자가 누구인지, 그 승자의 점수가 몇 점인지 출력까지 하면 이 문제를 해결할 수 있습니다!

```python
def appendDelete(s, t, k) :
    
    d = ""
    copy_t = t # t의 문자열을 복사한 copy_t
    
    for i in  range(0, k // 2 + 1) : # (k // 2 + 1)번 지워라 
        s = s[:-1] # 마지막 글자를 지우고 다시 s에 저장 → (k // 2 + 1) 횟수만큼 반복
    
    for i in range(0, k // 2) : # (k // 2)번 추가하라. 
        
        d += copy_t[-1:] # 역순으로 된 문자열 한글자씩 총 k // 2번 반복하여 k // 2개의 글자를 d에 저장!
        copy_t = copy_t[:-1] # 마지막 글자를 지우고 다시 copy_t에 저장 → (k // 2) 횟수만큼 반복
        
    reverse_d = d[::-1] # 역순으로 저장된 d를 다시 역순으로 즉, → 본래 순서로!
    
    if(s + reverse_d == t) :
        return "Yes"
    else :
        return "No"
```

```python
appendDelete("hackerhappy", "hackerrank", 9)
```

```python
appendDelete("aba", "aba", 7)
```

```python
appendDelete("ashley", "ash", 2)
```

이제 알파벳을 추가하는 반복문을 작성하기 전 원하는 알파벳으로 어떻게 추가를 해야할까 생각을 했습니다. 그래서 생각해 낸 큰 틀은 아래와 같습니다. 문자열 t의 가장 마지막 알파벳을 하나씩 꺼내어 다른 변수에 저장을 합니다. 그런데 스택과 비슷한 원리로 이 다른 변수에는 역순으로 저장이 됩니다. 따라서 이 변수에 저장된 문자열을 다시 역순으로 바꾸어준 뒤 t // 2 + 1만큼 알파벳을 삭제한 s 문자열과 더하면 됩니다. 그리고 이 더한 문자열이 t와 같은지 아닌지 if문을 이용한 검사를 통해 리턴 "Yes" 또는 "No"를 합니다. 
이제 세부적으로 위의 설명을 해보겠습니다.
t의 문자열은 기존의 문자열의 형태가 변하지 않아야 하므로 이 변수 t에 있는 문자열을 똑같이 복사한 copy_t를 선언하고 t의 문자열을 저장합니다. 그리고 반복문을 이용하는데 추가의 경우 k // 2의 개수만큼 알파벳을 추가하는 것에 주의하여 범위를 지정하고, 이 copy_t의 마지막 문자를 의미하는 copy_t[-1:]를 함수 최상단에 선언한 d 변수에 저장을 합니다(단 중첩 연산자로 누적 저장 되게끔 하기!). 그리고 변수 d에 이 copy_t의 마지막 문자 저장을 완료했기 때문에 copy_t를 마지막 문자를 지운 변수로 다시 초기화 합니다. 이렇게 k // 2번 만큼 반복문이 끝나면, 변수 d에는 우리가 문자열 s에 추가해야 할 문자열이 역순으로 저장되어 있습니다. 따라서 반복문이 모두 끝난 후 reverse_d라는 변수를 통해 d[::-1]를 해주면 역순 형태의 문자열을 다시 역순으로 즉, 본래의 순서로 저장이 됩니다. 이렇게 본래의 순서로 저장이 된 reverse_d와 s를 더한 문자열이 t와 같은지 아닌지를 if문을 이용하여 각각 리턴 값을 다르게하여 출력하면 문제 해결됩니다!
