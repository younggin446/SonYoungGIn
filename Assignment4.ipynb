#### 3번 문제

range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.

range(stop)
range(start, stop[, step])

2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.

3) frange 사용 예

3-1) frange(0.5)

[0.0, 0.1, 0.2, 0.3, 0.4]

3-2) frange(1.0, 2.0)

[1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]

3-3) frange(2.2, 4.0, 0.5)

[2.2, 2.7, 3.2, 3.7]

```python
def frange(*values) :
    list = []
    
    for i in values :
        if(i < 0) :
            print("음수를 받을 수 없습니다.")
            return
        
    if(len(values) == 1) :
        start = 0.0
        step = 0.1
        stop = values[0]
        
        while(True) :
            list.append(float("%.1f" %(1 * start)))
            
            start += step
            if ( start >= stop ):
                return list
            
    if(len(values) == 2) :
        start = values[0]
        stop = values[1]
        step = 0.1
        
        while(True) :
            list.append(float("%.1f" %(1 * start)))
            
            start += step
            if ( start >= stop ):
                return list
            
    if(len(values) == 3) :
        start = values[0]
        stop = values[1]
        step = values[2]
        
        while(True) :
            list.append(float("%.1f" %(1 * start)))
            
            start += step
            if ( start >= stop ):
                return list
```

```python
print(frange(-1.5))
print(frange(0.8))
print(frange(1.2, 2.5))
print(frange(1.2, 2.5, 0.3))
```

#### 설명

매개변수로 1개 ~ 3개까지 몇 개를 받을 지 모르기 때문에 frange함수의 매개변수로는 가변 인자를 받습니다. 그리고 각각의 실수를 저장할 빈 리스트를 생성합니다.
먼저 frange()의 인수들은 음수를 받지 않는다고 했으므로 매개변수로 전달받은 values들을 for 반복문을 이용하여 음수를 걸러냅니다.
매개변수로 만약 1개만 전달되면 stop이 전달 된 것이므로, start와 step은 각각 0.0과 0.1로 초기화하고 stop만 values의 0번째 인덱스의 값으로 초기화합니다. 이유는 stop전까지의 실수만 출력하기 위함입니다.
그리고 while문을 이용하여 0.0부터 stop전까지의 실수들을 리스트에 삽입합니다. 이 때 1자리의 실수를 삽입해야 하므로 %.1f로 범위를 지정하고 start에 1을 곱한 값을 삽입합니다.
start는 아래에서 step을 더한 값으로 다시 초기화 되기 때문에 while문을 돌면서 문제에서 요구하고자 하는 stop전까지의 소수들이 리스트에 삽입됩니다.
이렇게 삽입되면서 start의 값이 stop보다 크거나 같아질 때 리스트의 값을 리턴합니다.

매개변수로 만약 2개가 전달되면 start와 stop이 전달 되는 것이므로 step만 0.1로 초기화하고 start와 stop은 values의 0번째 인덱스와 1번째 인덱스의 값으로 초기화합니다.
역시 while문을 돌면서 start의 값을 step을 더한 값으로 다시 초기화하며 list에 삽입하면 start부터 stop전까지의 실수들이 리스트에 올바르게 삽입됩니다.

마지막 매개변수로 3개가 전달되면 values의 0, 1, 2번째 인덱스의 값을 start, stop, step으로 각각 초기화 하고 마찬가지로 while문을 돌면서 start의 값을 step과 더한 값으로 다시 초기화하며 list에 삽입하면 start부터 stop까지 step 간격으로 실수들이 올바르게 리스트에 삽입됩니다.

```python
def sum(*values) :
    
    sum = 0
    for i in values :
        sum += i
        
    return sum
```

```python
def myinitial(string) :
    string = string.split(" ")
    
    def return_first_char(s) :
        return s[0]
    
    string = list(map(return_first_char,string))
    
    string = ("".join(string).upper())
        
    print(string)
```

#### 6번 문제

음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

[옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.

```python
def factorial(n) :
    if n == 1 :
        return 1
    else :
        return n * factorial(n - 1)
```

```python
# 가장 많이 나온 단어와 개수 결과창 가장 하단에 존재
from urllib import request
import string
import pickle

f = request.urlopen("https://www.naver.com/")
source_bytes = f.read()
source = source_bytes.decode("UTF-8")
stp = string.punctuation

source = source.strip() # source 코드의 앞 뒤 공백 제거
exceptHTML_list = [] # HTML 태그 제외한 문자열들을 저장할 배열
word_flag = True # "<"와 ">" 사이의 문자열들을 제외한 나머지 문자열들을 exceptHTML_list 배열에 삽입하기 위해 필요한 flag 변수
punctuation = string.punctuation # 구두문자를 string 변수에 저장

for i in range(len(source)) :
    
     if(i <= len(source) -1):
        if(source[i] == "<" and source[i + 1:i + 7].lower() == "script") :
            j = i + 7

            while(True) :
                if(source[j] == "<" and source[j + 1] == "/" and source[j + 2 : j + 8].lower() == "script") :
                    source = source[:i] + source[j+9:]
                    break
                j = j + 1
                if(j > len(source)) :
                    break

for i in range(len(source)) :
    
     if(i <= len(source) -1):
        if(source[i] == "<" and source[i + 1:i + 6].lower() == "style") :
            j = i + 7

            while(True) :
                if(source[j] == "<" and source[j + 1] == "/" and source[j + 2 : j + 7].lower() == "style") :
                    source = source[:i] + source[j+7:]
                    break
                j = j + 1
                if(j > len(source)) :
                    break
    
for ch in source :
        
    if (ch == "<") :
        word_flag = False
        continue
        
    elif (ch == ">" and word_flag == False) :
        word_flag = True
        continue 
        
    if(word_flag) :
        exceptHTML_list.append(ch)


exceptHTML_list = ("".join(exceptHTML_list)).split()

# 문자열 끝에 있는 구두 문자를 기준으로 분리함 → 구두문자를 제외한 공백이 발생, 문자열 내에 있는 구두 문자는 지워지지 않음.
for i in range(len(exceptHTML_list)) :
    exceptHTML_list[i] = exceptHTML_list[i].strip(punctuation)

# 문자열 내에 있는 구두 문자 없애는 for문
for i in range(len(exceptHTML_list)) :
    for j in punctuation :
        if(j in exceptHTML_list[i]) :
            num = exceptHTML_list[i].find(j)
            exceptHTML_list[i] = exceptHTML_list[i][:num] + exceptHTML_list[i][num + 1 : ]

# 위에서 발생한 공백을 제거하기 위한 리스트 내포 함수
exceptHTML_list = [v for v in exceptHTML_list if v]
    
dic = dict() # 사전 타입의 dict 생성
for i in exceptHTML_list:
    dic[i] = 0 # 단어들의 개수를 세기 위함으로 모두 0으로 초기화
    
for i in dic.keys():
    for j in exceptHTML_list:
        if (j == i):
            dic[i]+=1
print(dic)

#불용어 제거
f = open('stopword.txt','r', encoding = "UTF-8")
s = f.read()
f.close()

s = s.split('\n')

lst = []

for i in dic.keys():
    lst.append(i)
    
for i in lst:
    for j in s:
        if ( i == j ):
            del dic[i]
            break
            
#단어의 출현빈도 사전 객체 저장
f = open('www.naver.com.words.frequency.txt','wb')
pickle.dump(dic, f)
f.close()

#가장 많은 단어의 출현 3개
f = open('www.naver.com.words.frequency.txt','rb')
diction = pickle.load(f)
many_word_list = []

for (i,j) in diction.items():
    many_word_list.append( (i,j) )
    
def first_index_word(lst):
    return lst[1]

many_word_list = sorted(many_word_list, key=first_index_word, reverse=True)

for i in range(3):
    print (many_word_list[i])
```
